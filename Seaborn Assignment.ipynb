```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the "fmri" dataset
df = sns.load_dataset("fmri")

# Plot line plot using "timepoint" as x-axis and "signal" as y-axis
sns.lineplot(x="timepoint", y="signal", data=df, hue="event", style="region")

# Set the title and labels
plt.title("FMRI Signal over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


df = sns.load_dataset("titanic")


plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=df)
plt.title('Box Plot of Age by Pclass')


plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=df)
plt.title('Box Plot of Fare by Pclass')


plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


df = sns.load_dataset("diamonds")


sns.histplot(data=df, x='price', hue='cut', multiple='stack')


plt.title("Histogram of Price with Cut")
plt.xlabel("Price")
plt.ylabel("Count")


plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
df = sns.load_dataset("flights")

# Pivot the data to create a matrix
pivot_data = df.pivot(index='month', columns='year', values='passengers')

# Plot the heatmap
sns.heatmap(data=pivot_data, cmap='YlGnBu')

# Set the title and labels
plt.title("Flights Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")

# Display the plot
plt.show()
```
